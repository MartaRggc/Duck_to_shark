# Columna 'age'

```python
# Vayamos a por esta columna a ver qué tal

print(shark2.age.unique()) 
shark2.age.isna().mean()*100
```

Vaya, tiene bastantes valores nulos (45%). Vemos tambiénque hay varios casos que aportan información imprecisa, 
como 'teens', 'eldery' o '(adult)'.
También hay casos que describen dos ataques a la vez, y casos en los que la edad está expresada en meses.
Identificaremos y gestionaremos todos los casos con una función basada en regex, que haga lo siguiente:

+ Devolver -1 si la información es ambigua (no numérica)
+ Devolver -1 si nos hablan de 'middle age'. 'mid-30s' también es muy ambiguo
+ Devolver 1 si la información está en meses
+ Devolver un entero redondeado con la media de las edades si la información es del tipo '13 or 14'
+ Devolver -1 si la información es sobre dos personas distintas, ya que son pocos casos y dificultarán obtener conclusiones sobre ellos a partir del resto de columnas

```python
shark2.age.fillna(-1, inplace = True)

def clean_age(x):
    if x != -1:
        if 'month' in x:
            res = 1
        elif 'mid' in x:
            res = -1
        elif 'or' in x:
            ages = re.findall('\d+', x)
            res0 = [int(e) for e in ages]
            res = int(np.mean(res0))
        else:
            ages = re.findall('\d+', x)
            if len(ages) == 1:
                res = int(ages[0])
            else:
                res = -1
    else:
        res = -1
            
    return res

shark2.age = [clean_age(x) for x in shark2.age]

shark2.age.unique()
```

# Columna 'injury'

Esta columna es un caos porque es la más descriptiva de todas. Sin embargo, podemos sacar información 
interesante de ella buscando palabras clave, y determinar si la herida en sí fue grave o leve.

Buscaremos esta información y la almacenaremos en una nueva columna llamada 'severity' por si luego queremos
relacionarla con la fatalidad y encontrar correlaciones con otros factores.

```python
# Echemos ahora un vistazo al tipo de herida, rellenando los nulos antes (que son muy pocos) con 'unknown':

shark2.injury.fillna('unknown',inplace = True)
shark2.injury.unique()[:20]
```

```python
# Echando un ojo a las descripciones que contienen 'no' + 'injury','minor' o 'slight', 
# observamos que corresponden a ataques sin daño personal o con daños menores:


for i in shark2.injury.unique():
    if 'no' in i.lower() and 'injury' in i.lower() or 'minor' in i.lower() or 'slight' in i.lower():
        print(i)
    else: 
        pass
```

```python
# Creamos ahora la columna 'severity' y la ponemos entre 'injury' y 'fatality':

col_names = ['original_order',
       'day', 'month','year','time', 'country', 'area', 'location',
       'activity', 'name', 'sex', 'age', 'type', 'injury', 'fatal', 'species',
       'investigator_or_source', 'pdf', 'href_formula', 'href','date', 'unn1','unn2','severity']

shark2.columns = col_names

shark2 = shark2.reindex(columns = ['original_order',
       'day', 'month','year','time', 'country', 'area', 'location',
       'activity', 'name', 'sex', 'age', 'type', 'injury','severity', 'fatal', 'species',
       'investigator_or_source', 'pdf', 'href_formula', 'href','date', 'unn1','unn2'])

shark2.head()
```

```python
# Ahora la vamos rellenando en función de la información de 'fatal' y 'injury'. Habrá tres grados de gravedad,
# 0, 1 y 2 yendo de menos a más grave:

shark2.severity = [2 if i == 1 else 0 for i in shark2.fatal] # Si ha habido muertes la herida ha sido grave

def sever(j):
    if 'no ' in j.lower() and 'injury' in j.lower(): # Primero buscamos casos sin daños humanos
        return 0
    elif 'minor' in j.lower() or 'slight' in j.lower() or 'shallow' in j.lower(): # Luego daños menores
        return 1
    else: # Lo que quede serán daños graves a personas, en su mayoría
        return 2 

    
# Aplicamos la función a los casos que no hayamos etiquetado ya como graves:

shark2.severity = [sever(j) if i != 2 else 2 for i,j in zip(shark2.severity,shark2.injury)] 

shark2.head() # ¡Terminado!
```

# Columna 'activity'

```python
# Echemos un primer ojo a ver qué tipo de valores tiene esta columna:

shark2.activity.value_counts()[shark2.activity.value_counts()>15]
```

Bueno, vemos que hay muchas categorías que se podrían unificar. Vamos a resumirlas en las siguientes:

+ surfing (includes body boarding, boogie boarding and body surfing)
+ swimming (includes snorkeling, bathing and falling from boats)
+ fishing (includes spearfishing)
+ diving (includes scuba diving)
+ wading (includes standing and walking)
+ kayaking
+ other


```python
keywords = {'surfing':['boarding','surf'],
          'swimming':['swim','snorkel','bathing','fell','fall','boat','rescue','river'],
          'fishing':['fishing','spear','fish','lobster','shrimp'],
          'diving':['dive','diving','scuba'],
          'wading':['stand','wading'],
          'kayaking':['kayak','canoe']}

def checkactivity(x):
    
    res2 = 0
    
    try:
        for i in keywords.keys():
            for j in keywords[i]:
                if j in x.strip().lower():
                    res2 = i
                else:
                    pass
        if res2 == 0:
            return 'other'
        else:
            return res2
    except:
        return x

shark2.activity = [checkactivity(x) for x in shark2.activity]
shark2.activity.fillna('other',inplace = True)

# Veamos lo que queda:
shark2.activity.value_counts()
```

```python
# Hay un porrón. Vamos a ir agrupando en los más típicos, y el resto los pondremos como 'other'.

keywords = ['lemon','white','whitetip','hammer','whaler','blue','mako','bull',
           'tiger','blacktip','nurse','raggedtooth','wobbegong','basking',
           'sandtiger','spinner','reef','zambesi','angel']

for i in range(shark2.species.shape[0]):
    c = 0
    for j in keywords:
        if j in shark2.species.iloc[i].lower():
            shark2.species.iloc[i] = j
            c +=1  
    if c == 0:
        if 'unconfirmed' in shark2.species.iloc[i].lower():
            shark2.species.iloc[i] = 'shark presence unconfirmed'
        else:
            shark2.species.iloc[i] = 'other'

shark2.species.value_counts()
```
