Ahora que está más limpia, vamos a intentar convertir los datos directamente a formato datetime, teniendo en cuenta los casos más típicos en los que tenemos estructura día-mes-año, mes-año ó directamente año. Para ello
definimos la función checkdate:

```python
def checkdate(x):
    try:
        return pd.to_datetime(x,format = "%d-%b-%Y") # Fechas completas
    except:
        try:
            return pd.to_datetime(x,format = "%b-%Y") # Fechas con mes y año
        except:
            try:
                return pd.to_datetime(x,format = "%Y") # Fechas con solo el año
            except:
                try:
                    return pd.to_datetime(x,format = "%d.%m.%Y") # Por si alguna fecha está en formato tipo 04.06.1994
                except:
                    try:
                        return pd.to_datetime(x,format = "%Y.%m.%d") # O del revés
                    except:
                        try:
                            return pd.to_datetime(x,format = "%Y.%m") # O del revés
                        except:
                            return x

shark.date = [checkdate(i) for i in shark.date]
```

```python
# Echamos un ojo a las fechas que no han podido convertirse a formato datetime:

for i in shark.date:
    if not isinstance(i, pd.Timestamp):
        print(i)
```

+ Cambiamos 'july' por 'jul', 'april' por 'apr', etc.
+ Quitamos las 's' que están ocultando algunos años del tipo 1960s
+ Quitamos los .a, .b, .c, .r etc.

Observamos que de las fechas que quedan sin pasar a datetime falta información también en las columnas year
y time. Algunas no se han pasado a formato pd.Timestamp porque son años anteriores a 1677, que es el límite de 
antiguedad de fecha para este formato. Para arreglarlo, haremos lo siguiente:

En vez de tener una columna 'date' con la información compuesta de año, mes y día del mes, dividiremos esta información en tres columnas, ya que tenemos otras vacías para poder sobreescribir y ya hay una (year) en la que tenemos el año.

Seguiremos estos pasos:

1) Mirar en qué columnas de year falta el año o es cero, y sustituir por el valor de date si lo hay. Si no se puede, marcamos con -1 dicho campo.
2) Renombrar una de las columnas sobrantes con 'month', y sacar la información de date si la hay en forma numérica. Si no, dejamos -1.
3) Lo mismo con la columna nueva columna 'day'.


```python
# Para el primer caso creamos una función:

def case1(i,j):  # donde i será el elemento de shark.year y j el de shark.date
    
    if i == np.NaN or i == 0: # comprobamos si el elemento es nulo o 0
        try:
            return int(j.year) # buscamos el año en la columna shark.date y lo asignamos como entero
        except:
            return 'unknown' # si no podemos sacarlo asignamos desconocido
        
    else: # si el elemento de shark.year no es nulo ni cero
        try:
            return int(i) # lo dejamos como está pasándolo a entero
        
        except: # si no podemos pasarlo a entero es que algo le pasa, intentamos sacarlo de date
            try:
                return int(j.year) # buscamos el año en la columna shark.date y lo asignamos como entero
            except:
                return i # si no podemos sacarlo lo dejamos como está y ya miraremos qué le pasa luego
            
shark.year = [case1(i,j) for i,j in zip(shark.year,shark.date)]

shark.year.unique()
```

Vemos que hay muchos con b.c., otros con intervalos entre años (1890-1891) etc. Vamos a usar un patrón de
regex para extrar un poco más de información del dataframe original, y los que no identifiquemos los dejaremos con
año desconocido 'unknown'. Creamos una minifunción para ello:

```python
def final_year(x):
    try:
        if 'b.c' in x.lower() or 'bc.' in x.lower():
            a = -1 # Los que sean antes de cristo los expresaremos como números negativos
        else:
            a = 1
            
        ages = re.findall('\d+', x) # buscamos los números que haya
        
        if ages:
            res0 = [int(e) for e in ages]
            res = a*int(np.mean(res0)) # Haremos la media de los números que encontremos
        else:
            res = 'unknown' # si no encontramos números, asignamos desconocido
    except:
        res = 'unknown' # si da error, lo mismo
    
    return res

shark.year = [final_year(i) if j == 'unknown' else j for i,j in zip(shark_ori.Date,shark.year)]


# Echemos un ojo a lo que queda y veamos qué era en el dataframe original:

shark_ori.loc[shark[shark.year == 'unknown'].index] 
```

```python
# Son unos pocos casos con registros de fecha muy imprecisos. Directamente borramos las filas, ya que
# son muy pocos datos y no resultarán significativos en un análisis estadístico:

shark.drop(index = shark[shark.year == 'unknown'].index,inplace  = True)

shark.year.unique() # Columna year limpia y lista
```

# Columna 'time'

```python
# Vamos a por los valores nulos.

print(shark.time.unique()) 
print(shark.time.value_counts())
shark.time.isna().mean()*100
```

Esta columna tiene muchos valores nulos (el 53% casi), y tiene también muchos valores únicos diferentes. La pasaremos a formato pd.Timestamp como la columna date, y luego distribuiremos las horas en franjas del día.

```python
# Rellenamos los nulos con 'unknown', ya que esta columna será tipo string cuando esté limpia:

shark.time.fillna('unknown', inplace = True)

# Definimos cuatro franjas del día:

daytimes = {'morning':[pd.to_datetime('06h00',format = "%Hh%M"),pd.to_datetime('12h00',format = "%Hh%M")],
          'afternoon':[pd.to_datetime('12h01',format = "%Hh%M"),pd.to_datetime('18h00',format = "%Hh%M")],
          'evening':[pd.to_datetime('18h01',format = "%Hh%M"),pd.to_datetime('21h00',format = "%Hh%M")],
          'night':[pd.to_datetime('21h01',format = "%Hh%M"),pd.to_datetime('05h59',format = "%Hh%M")]}

keywords = {'morning':['morning','am','a.m.','dawn','daybreak','daytime'],
          'afternoon':['afternoon','midday','after noon','lunch','noon'],
          'evening':['evening','dusk','sunset','sundown','p.m.'],
          'night':['night','midnight','dark']}


# Primero vamos a distribuir aquellos valores localizables con formato datetime:
# Buscamos con un patrón de regex, ya que hay muchos casos en los que hay varias horas, y nos quedaremos con la
# primera de ellas:

def checktime(x):
    try:
        pattern = r'\d\dh\d\d'
        res0 = re.findall(pattern,x)
                
        if res0: # si encontramos algo en este formato de horas
            res = pd.to_datetime(res0[0],format = "%Hh%M") # horas normales
            for i in daytimes.keys():
                if res > daytimes[i][0] and res <= daytimes[i][1]:
                    return i
                else:
                    pass
                
        else: # si no, y solo es una string:
            
            res2 = 0
            for i in keywords.keys():
                for j in keywords[i]:
                    if j in x.lower():
                        res2 = i
                    else:
                        pass
                    
            if res2 == 0:
                return x
            else:
                return res2
    except:
        return x.lower()

shark.time = [checktime(x) for x in shark.time]

# Veamos lo que queda:
shark.time.unique()
```

```python
# Quedan pocos valores distintos. La columna está lo suficientemente limpia, por lo que al resto de valores 
# les asignamos unknown y lo damos por terminado:

shark.time[(shark.time != 'morning')&(shark.time != 'afternoon')&(shark.time != 'evening')\
           &(shark.time != 'night')] = 'unknown'

shark.time.value_counts()
```
